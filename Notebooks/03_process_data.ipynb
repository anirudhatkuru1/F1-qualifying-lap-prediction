```python
import pandas as pd

df = pd.read_csv("../data/all_quali_2024.csv")

# -------------------------------------
# 1. Basic cleaning
# -------------------------------------

# Check missing values
print(df.isnull().sum())

# Remove duplicate rows
df = df.drop_duplicates()

# Keep only laps with valid lap_time
df = df.dropna(subset=["lap_duration"])

# Convert datatypes
df["lap_number"] = pd.to_numeric(df["lap_number"], errors="coerce")
df["duration_sector_1"] = pd.to_numeric(df["duration_sector_1"], errors="coerce")
df["duration_sector_2"] = pd.to_numeric(df["duration_sector_2"], errors="coerce")
df["duration_sector_3"] = pd.to_numeric(df["duration_sector_3"], errors="coerce")
df["lap_duration"] = pd.to_numeric(df["lap_duration"], errors="coerce")

# -------------------------------------
# 2. Add derived features
# -------------------------------------

# Total sector time sum (should equal lap duration but sometimes slight diff)
df["sector_sum"] = (
    df["duration_sector_1"] +
    df["duration_sector_2"] +
    df["duration_sector_3"]
)

# Average speeds combined
df["avg_speed_est"] = (df["i1_speed"] + df["i2_speed"] + df["st_speed"]) / 3

# Identify pit-out laps (model should not train on these)
df["is_pit_out_lap"] = df["is_pit_out_lap"].astype(int)

# -------------------------------------
# 3. Remove pit-out laps (optional but recommended)
# -------------------------------------

df = df[df["is_pit_out_lap"] == 0]

# -------------------------------------
# 4. Save processed data
# -------------------------------------

df.to_csv("../data/processed_quali_2024.csv", index=False)
print("✅ Processed data saved to ../data/processed_quali_2024.csv")

df.head()
```
