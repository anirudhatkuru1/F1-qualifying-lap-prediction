```python
import pandas as pd

df = pd.read_csv("../data/all_quali_2024.csv")
print(df.shape)
df.head()
```

```python
# Check missing values
df.isnull().sum()

# Check duplicate rows
df.duplicated().sum()

df = df.drop_duplicates()

df = df.dropna(subset=["lap_duration"])  # keep only rows with lap times

# Example: standardize driver names
df["driver"] = df["driver"].str.strip().str.title()

# Standardize team names
df["team"] = df["team"].str.strip().str.upper()

# Add lap number as numeric
df["lap_number"] = pd.to_numeric(df["lap_number"], errors="coerce")

# Compute cumulative average lap time per driver in that session
df["driver_avg_lap"] = df.groupby(["session_key", "driver"])["lap_duration"].transform("mean")

# Encode tire compounds 
df = pd.get_dummies(df, columns=["tire_compound"])

df.to_csv("../data/processed_quali_2024.csv", index=False)
print("âœ… Processed data saved to ../data/processed_quali_2024.csv")


```
