```python
import requests
import pandas as pd

# Load session keys with year from the CSV saved above
df_sessions = pd.read_csv("../data/all_qualifying_sessions_2015_2024.csv")
session_keys = df_sessions["session_key"].tolist()

all_laps = []

for key in session_keys:
    print(f"Fetching laps for session {key}...")
    url = "https://api.openf1.org/v1/laps"
    params = {"session_key": key}
    
    response = requests.get(url, params=params)
    laps = response.json()
    
    if laps:
        for lap in laps:
            lap["session_key"] = key
            # add year info too by matching session_key with session dataframe
            lap["year"] = int(df_sessions[df_sessions["session_key"] == key]["year"].values[0])
        all_laps.extend(laps)

df_all_laps = pd.DataFrame(all_laps)
print(f"Total laps fetched: {len(df_all_laps)}")

# Save combined CSV
df_all_laps.to_csv("../data/all_quali_2015_2024.csv", index=False)
print("âœ… Saved all qualifying laps 2015-2024 to ../data/all_quali_2015_2024.csv")
```
