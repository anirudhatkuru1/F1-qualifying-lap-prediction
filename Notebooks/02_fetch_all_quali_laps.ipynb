```python
import requests
import pandas as pd
from pathlib import Path
```

```python
# Replace these with your session_keys
session_keys = [9689, 9994, 10002, 10010, 10018]  # all 2024 qualifying sessions

all_laps = []

for key in session_keys:
    print(f"Fetching session {key}...")
    url = "https://api.openf1.org/v1/laps"
    params = {"session_key": key}
    response = requests.get(url, params=params)
    laps = response.json()
    
    # Add session_key to each lap row
    for lap in laps:
        lap["session_key"] = key
    
    all_laps.extend(laps)

# Convert to DataFrame
df_all_laps = pd.DataFrame(all_laps)
print(f"Total laps fetched: {len(df_all_laps)}")

# Save combined CSV
df_all_laps.to_csv("../data/all_quali_2024.csv", index=False)
print("✅ Saved to ../data/all_quali_2024.csv") 
```
