```python
import requests
import pandas as pd

all_sessions = []

for year in range(2015, 2025):  # adjust the range as needed
    print(f"Fetching qualifying sessions for {year}...")
    url = "https://api.openf1.org/v1/sessions"
    params = {"year": year, "session_name": "Qualifying"}
    
    response = requests.get(url, params=params)
    data = response.json()
    
    if data:
        for session in data:
            session['year'] = year  # add the year to each session
        all_sessions.extend(data)

df_sessions = pd.DataFrame(all_sessions)

if df_sessions.empty:
    print("⚠️ No qualifying sessions found.")
else:
    cols_to_show = [c for c in ["year", "meeting_name", "session_name", "session_key"] if c in df_sessions.columns]
    display(df_sessions[cols_to_show].head())

# Save to CSV for use in next steps
df_sessions.to_csv("../data/all_qualifying_sessions_2015_2024.csv", index=False)
print("✅ Saved all qualifying sessions for 2015-2024 to ../data/all_qualifying_sessions_2015_2024.csv")
```
